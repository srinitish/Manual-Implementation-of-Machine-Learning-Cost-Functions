```python
import math
y_true = [
    [0,1,0],
    [0,0,1]
]
```

```python
y_pred = [
    [0.1,0.6,0.3],
    [0.2,0.2,0.6]
]
```

```python
def category_entropy(y_true,y_pred):
    n = len(y_true)
    total_loss =0
    for i in range(n):
        sample_loss =0
        for j in range(len(y_true[i])):
            sample_loss+= y_true[i][j]*math.log(y_pred[i][j])
        sample_loss = -sample_loss
        total_loss+=sample_loss
    loss = total_loss/n
    return loss
        
```

```python
category_entropy(y_true,y_pred)
```
