```python
import numpy as np
y_true = [1,2,0]
y_pred = [[0.2,0.7,0.3],[0.5,0.1,0.8],[0.6,0.3,0.1]]
```

```python
def sparse_entropy(Y_true,y_pred):
    loss = 0
    n=len(y_true)
    for i in range(n):
        true_class = y_true[i]
        predict = y_pred[i][true_class]
        loss+=-np.log(predict)
    return loss/n
```

```python
sparse_entropy(y_true,y_pred)
```
